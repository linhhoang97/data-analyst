```python
import pandas as pd
```

```python
data = pd.read_csv('Dummy Data for Technical Test.csv')
data
```

```python
data = data.dropna(subset = ['DoB', 'Gender', 'City'])
data
```

```python
 pd.options.mode.chained_assignment = None
data['DoB'] = pd.to_datetime(data['DoB'], format= '%Y-%m-%d', errors = 'coerce')
```

```python
from datetime import datetime
```

```python
data['Age'] = datetime.now().year - pd.DatetimeIndex(data['DoB']).year  
```

```python
data['Age'] = data['Age'].fillna('0').astype(int)
```

```python
data
```

```python
ages = data.iloc[:, -1].values
ages
```

```python
import numpy as np
median = np.median(ages)
print('\n median \n', median)
```

```python
cutoff = median * 3 #k-value = 3
cutoff
```

```python
lower, upper = median - cutoff, median + cutoff
upper, lower
```

```python
mylist = []

for age in ages:
    if age == 0 or age > upper or age < lower:
        mylist.append(median)
    else:
         mylist.append(age)
```

```python
mylist
```

```python

data['Age'] = mylist
data
```
